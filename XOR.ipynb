```python
#Deep Learning on XOR logic
import tensorflow as tf
import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Define Dataset
x_train=np.array([[0,0],[0,1],[1,0],[1,1]],"float32")
y_train=np.array([[0],[1],[1],[0]],"float32")
```

```python
#Build NN
xor=tf.keras.models.Sequential()
xor.add(tf.keras.layers.Dense(4, input_dim=2, activation='relu'))
xor.add(tf.keras.layers.Dense(1, activation='sigmoid'))

#Compile
xor.compile(optimizer='adam', loss='mean_squared_error', metrics=['binary_accuracy'])
```

```python
#TRAIN
result=xor.fit(x_train, y_train, epochs=500, verbose=2)
#TEST
print(xor.predict(x_train).round())
print(xor.predict([[1,0]]).round())
```

```python
#Loss and Accuracy Data Collection
loss_curve=result.history['loss']
accu_curve=result.history['binary_accuracy']

plt.plot(loss_curve, label='Train')
plt.show()

plt.plot(accu_curve, label='Train')
plt.show()
```
